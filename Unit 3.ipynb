# Violin Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("tips")
```

```python
df
```

```python
sns.violinplot(x="day", y="total_bill", data=df)
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=df, hue="time")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=df, hue="time", linewidth = 2)
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=df, hue="time", linewidth = 2, palette = "cool")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=df, hue="time", linewidth = 2, palette = "jet_r")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=df, hue="time", linewidth = 2, palette = "vlag")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=df, hue="time", linewidth = 2, palette = "turbo_r")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=df, hue="time", linewidth = 2, palette = "plasma")
plt.show()
```

```python
sns.violinplot(x="time", y="total_bill", data=df, order = ["Dinner","Lunch"], palette = "cool")
plt.show()
```

```python
sns.violinplot(x="time", y="total_bill", data=df, order = ["Dinner","Lunch"], color = 'r', saturation = 0.4)
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=df, hue = "sex", split = True)
plt.show()
```

```python
sns.violinplot(y="day", x="total_bill", data=df, hue = "sex", split = True)
plt.show()
```

```python
sns.violinplot(x="time", y="total_bill", data=df, order = ["Dinner","Lunch"], inner = "quart2")
plt.show()
```

```python
sns.violinplot(x="time", y="total_bill", data=df, order = ["Dinner","Lunch"], inner = "box")
plt.show()
```

```python
sns.violinplot(x="time", y="total_bill", data=df, order = ["Dinner","Lunch"], inner = "point")
plt.show()
```

```python
sns.violinplot(x="time", y="total_bill", data=df, order = ["Dinner","Lunch"], inner = "stick")
plt.show()
```

```python
sns.violinplot(x="total_bill", data=df)
plt.show()
```

```python
sns.violinplot(y="total_bill", data=df)
plt.show()
```

# Pair Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.pairplot(df, hue ="sex")
plt.show()
```

```python
sns.pairplot(df, vars = ["total_bill","tip"], hue ="sex")
plt.show()
```

```python
sns.pairplot(df, vars = ["total_bill","tip"], hue ="sex", hue_order = ["Female","Male"])
plt.show()
```

```python
sns.pairplot(df, vars = ["total_bill","tip"], hue ="sex", hue_order = ["Female","Male"], palette = "gist_rainbow_r")
plt.show()
```

```python
sns.pairplot(df, x_vars = ["total_bill","tip"], hue ="sex", hue_order = ["Female","Male"], palette = "gist_rainbow_r")
plt.show()
```

```python
sns.pairplot(df, y_vars = ["total_bill","tip"], hue ="sex", hue_order = ["Female","Male"], palette = "gist_rainbow_r")
plt.show()
```

```python
sns.pairplot(df, y_vars = ["total_bill","tip"], hue ="sex", hue_order = ["Female","Male"], palette = "gist_rainbow_r",
             kind = 'reg')
plt.show()
```

```python
sns.pairplot(df, y_vars = ["total_bill","tip"], hue ="sex", hue_order = ["Female","Male"], palette = "gist_rainbow_r",
             kind = 'scatter')
plt.show()
```

```python
sns.pairplot(df, y_vars = ["total_bill","tip"], hue ="sex", hue_order = ["Female","Male"], palette = "gist_rainbow_r",
             kind = 'hist')
plt.show()
```

```python
sns.pairplot(df, y_vars = ["total_bill","tip"], hue ="sex", hue_order = ["Female","Male"], palette = "gist_rainbow_r",
             kind = 'kde')
plt.show()
```

```python
sns.pairplot(df, hue ="sex", hue_order = ["Female","Male"], palette = "ocean", kind = 'kde', diag_kind = 'hist')
plt.show()
```

# Strip Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df
```

```python
sns.stripplot(x="day", y="total_bill", data=df)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", palette="autumn")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", palette="afmhot_r", linewidth = 2)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", palette="afmhot_r", linewidth = 2, edgecolor = 'r')
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", palette="afmhot_r", linewidth = 2, edgecolor = 'r', jitter=2)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", palette="afmhot_r", linewidth = 2, edgecolor = 'r', jitter=2, size=3)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", palette="afmhot_r", linewidth = 1, edgecolor = 'r', jitter=2, marker = "<", alpha = 0.5)
plt.show()
```
