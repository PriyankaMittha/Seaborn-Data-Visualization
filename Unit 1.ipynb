# Bar Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = sns.load_dataset("penguins")
```

```python
df1
```

```python
sns.barplot(x=df1.island, y=df1.bill_length_mm, palette='Accent')
plt.show()
```

```python
sns.barplot(x=df1.island, y=df1.bill_length_mm, palette='Accent', data = df1)
plt.show()
```

```python
sns.barplot(x=df1.island, y=df1.bill_length_mm, data = df1, hue = "sex", palette='rocket_r')
plt.show()
```

```python
sns.barplot(x=df1.island, y=df1.bill_length_mm, data = df1, hue = "sex", 
            palette='rocket_r', order = ['Dream','Torgersen','Biscoe'])
plt.show()
```

```python
sns.barplot(x=df1.island, y=df1.bill_length_mm, data = df1, hue = "sex", 
            palette='rocket_r', order = ['Dream','Torgersen','Biscoe'],
            hue_order = ["Female","Male"])
plt.show()
```

```python
sns.barplot(x=df1.island, y=df1.bill_length_mm, data = df1, hue = "sex", 
            palette='rocket_r', order = ['Dream','Torgersen','Biscoe'],
            hue_order = ["Female","Male"], errorbar=('ci', 15))
plt.show()
```

```python
sns.barplot(x="bill_length_mm", y="bill_depth_mm", data = df1, orient = 'h', palette = "Accent")
plt.show()
```

```python
sns.barplot(x=df1.island, y=df1.bill_length_mm, data = df1, hue = "sex", 
            palette='rocket_r', order = ['Dream','Torgersen','Biscoe'],
            hue_order = ["Female","Male"], saturation = 0.5, 
            err_kws={'color': 'r','linewidth': 10})
plt.show()
```

```python
sns.set_style(style="darkgrid")
sns.barplot(x=df1.island, y=df1.bill_length_mm, data = df1, hue = "sex", 
            palette='rocket_r', order = ['Dream','Torgersen','Biscoe'],
            hue_order = ["Female","Male"])
plt.show()
```

```python
sns.set_style(style="dark")
sns.barplot(x=df1.island, y=df1.bill_length_mm, data = df1, hue = "sex", 
            palette='rocket_r', order = ['Dream','Torgersen','Biscoe'],
            hue_order = ["Female","Male"])
plt.show()
```

# Histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = sns.load_dataset("penguins")
```

```python
df1
```

```python
sns.displot(df1['flipper_length_mm'])
plt.show()
```

```python
sns.displot(df1['flipper_length_mm'], bins = [170,180,190,200,210,220,230])
plt.show()
```

```python
sns.displot(df1['flipper_length_mm'], bins = [170,180,190,200,210,220,230], kde = True)
plt.show()
```

```python
sns.displot(df1['flipper_length_mm'], bins = [170,180,190,200,210,220,230], 
            kde = True, rug = True)
plt.show()
```

```python
sns.displot(df1['flipper_length_mm'], bins = [170,180,190,200,210,220,230], 
            kde = True, rug = True, color = 'r')
plt.show()
```

```python
sns.displot(df1['flipper_length_mm'], color = 'r', log_scale = True)
plt.show()
```
