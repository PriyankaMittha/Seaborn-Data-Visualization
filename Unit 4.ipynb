# Box Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("tips")
```

```python
df
```

```python
sns.boxplot(x="day", y="total_bill", data=df)
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=df, hue="time")
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=df, hue="time", color = 'y')
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=df, hue="sex", palette='dark:y', order = ['Fri','Sun','Thur','Sat'])
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=df, hue="sex", palette='dark:y', order = ['Fri','Sun','Thur','Sat'], showmeans = True)
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=df, hue="sex", palette='dark:y', order = ['Fri','Sun','Thur','Sat'], showmeans = True, meanprops = {'marker':"+", 'markeredgecolor':"b"})
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=df, hue="sex", order = ['Fri','Sun','Thur','Sat'], showmeans = True, meanprops = {'marker':"+", 'markeredgecolor':"b"}, palette="cubehelix_r")
plt.show()
```

```python
sns.boxplot(data=df, showmeans = True, meanprops = {'marker':"+", 'markeredgecolor':"b"}, palette="cubehelix_r")
plt.show()
```

```python
sns.boxplot(data=df, showmeans = True, meanprops = {'marker':"+", 'markeredgecolor':"b"}, palette="cubehelix_r", orient="h")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(y=df["total_bill"])
plt.show()
```

# Styling Plot in Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("tips")
```

```python
sns.set_style("dark")
sns.boxplot(x="day", y="total_bill", data=df)
plt.show()
```

```python
sns.set_style("darkgrid")
sns.boxplot(x="day", y="total_bill", data=df)
plt.show()
```

```python
sns.set_context("poster", font_scale = 2)
sns.boxplot(x="day", y="total_bill", data=df)
plt.show()
```

```python
sns.boxplot(x="sex", y="total_bill", data=df)
plt.show()
```

# Facet Grid

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("tips").head(20)
```

```python
sns.set_context("notebook")
fg = sns.FacetGrid(df, col = "sex", hue="day")
fg.map(plt.scatter, "total_bill", "tip")
plt.show()
```

```python

fg = sns.FacetGrid(df, col = "sex", hue="day")
fg.map(plt.bar, "total_bill", "tip")
plt.show()
```
